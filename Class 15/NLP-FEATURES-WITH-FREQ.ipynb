```python
import nltk
```

```python
from nltk.corpus import movie_reviews
len(movie_reviews.fileids())
```

```python
documents = []
for category in movie_reviews.categories():
    for fileid in movie_reviews.fileids(category):
        documents.append((movie_reviews.words(fileid), category))
documents[0:5]
```

```python
import random
random.seed(0)
random.shuffle(documents)

training_documents = documents[0:1500]
testing_documents = documents[1500:]
```

```python
from nltk.corpus import stopwords
stop = stopwords.words('english')
stop
```

```python
import string
punc = list(string.punctuation)
all_words = []
stop = stop + punc
for doc in training_documents:
    for w in doc[0]:
        if w.lower() not in stop:
            all_words.append(w.lower())
```

```python
all_words
```

```python
import nltk
dist = nltk.FreqDist(all_words)
features = dist.most_common(3000)
feature_words = [i[0] for i in features]
```

```python
feature_words
```

```python
from collections import Counter
    
def get_features(document):    
    features = {}
    for w in feature_words:
        if w in document:
            if w not in features:
                features[w] = 1
            else:
                features[w] += 1
                
    return features
    
    #features = dict(Counter(document))
    #return features
```

```python
get_features(training_documents[0][0])
```

```python
training_data = [(get_features(a), b) for (a, b) in training_documents]
```

```python
testing_data = [(get_features(a), b) for (a, b) in testing_documents]
```

```python
classifier = nltk.NaiveBayesClassifier.train(training_data)
```

```python
classifier.show_most_informative_features(10)
```

```python
nltk.classify.accuracy(classifier, testing_data)
```

```python
from nltk.classify.scikitlearn import SklearnClassifier
from sklearn.naive_bayes import MultinomialNB
multi_nb_classifier_sklearn = SklearnClassifier(MultinomialNB())
multi_nb_classifier_sklearn.train(training_data)
```

```python
nltk.classify.accuracy(multi_nb_classifier_sklearn, testing_data)
```
