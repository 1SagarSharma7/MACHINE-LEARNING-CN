```python
import pandas as pd
from nltk.tokenize import word_tokenize
```

```python
tweets_data = pd.read_csv("https://ninjasfiles.s3.amazonaws.com/attachment9844a9a7221e2e5b971f972051629676e0d2.csv")
```

```python
tweets_data
```

```python
tweets = tweets_data[:]['text']
```

```python
sentiments = tweets_data[:]['airline_sentiment']
```

```python
all_data = []
for i in range(len(tweets)):
    tweet = tweets[i]
    sentiment = sentiments[i]
    all_data.append((word_tokenize(tweet),sentiment)) 
```
