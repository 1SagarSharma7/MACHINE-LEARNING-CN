```python
def gd_runner(X_train, Y_train, learning_rate, num_iterations):
    
    coeff = np.zeros(14)
    
    for i in range(num_iterations):
        cur_cost = cost_gd(coeff,X_train, Y_train)
        print(i, " ",cur_cost)
        coeff = step_gradient(X_train, Y_train, learning_rate, coeff)
    
    return coeff
```
